## Key Observations
- No Augmentation: Strong train-test performance and highest training correlation.

- Augmentation v1 & v2: Noticeable increase in error and lower correlation on train, but similar or slightly improved test r (v1), suggesting stronger regularization (but possible underfitting).

- Augmentation v3: Best overall generalization—test error drops, and r improves over no augmentation.

## Probability Explanation
For each training batch, a random number (0–1) is generated and compared to the probability
- Probabilities introduce controlled randomness to augmentations:
- 0.0: Never applied
- 1.0: Always applied
- 0.5: Applied to about half of the samples
