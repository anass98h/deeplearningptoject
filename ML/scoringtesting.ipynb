```python

import os, random, pickle, math
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.model_selection import train_test_split, KFold, ParameterGrid
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr

from tensorflow.keras import backend as K
from tensorflow.keras.layers import (Input, SeparableConv1D, LayerNormalization,
                                     Activation, Add, GlobalAveragePooling1D,
                                     Dense, Dropout)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (EarlyStopping, ReduceLROnPlateau)

# ─── Reproducibility ───────────────────────────────────────────────────────
RANDOM_STATE = 42
os.environ['PYTHONHASHSEED']       = str(RANDOM_STATE)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

# ─── Configuration ─────────────────────────────────────────────────────────
# !! EDIT these two if your data lives elsewhere
SCORE_PATH    = 'data/dataset_for_scoring.csv'
SEQUENCE_DIR  = 'data/all_kinect_trimmed'

FRAMES      = 48
TEST_SIZE   = 0.10             # 10 % hold-out
CV_SPLITS   = 5                # 5-fold cross-val on the remaining 90 %
EPOCHS      = 300
ES_PATIENCE = 25

# only ~12 k params even at filters=32, n_features≈70
param_grid = {
    'filters'     : [32],          # conv channels
    'dense_units' : [(64, 32)],    # head
    'dropout'     : [0.10],
    'lr'          : [1e-3],
    'batch_size'  : [16],          # smaller batch for small data
}

# ─── Data Loading & Preprocessing ──────────────────────────────────────────
def load_data(score_csv, seq_dir):
    """Return list(X), array(y) after dropping the 'A1' outlier and scaling y to [0,4]."""
    print(">> Loading data …")
    df = pd.read_csv(score_csv)
    df = df[df['filename'] != 'A1']
    scores = df['Var2'].to_numpy().reshape(-1, 1)

    scaler_y = MinMaxScaler(feature_range=(0, 4))
    y = scaler_y.fit_transform(scores).ravel()

    X = []
    for fname in df['filename']:
        x = pd.read_csv(Path(seq_dir, f'{fname}_kinect.csv')).drop(columns='FrameNo').to_numpy()
        X.append(x)

    Path('models').mkdir(exist_ok=True)
    with open('models/score_minmaxscaler.pkl', 'wb') as f:
        pickle.dump(scaler_y, f)

    print(f"   → {len(X)} sequences loaded  (targets scaled to [0,4])\n")
    return X, y

def sample_sequence(seq, n_frames=FRAMES):
    """Uniformly sample or zero-pad sequence to n_frames."""
    T, D = seq.shape
    if T >= n_frames:
        idx = np.linspace(0, T - 1, n_frames).round().astype(int)
        return seq[idx]
    pad = np.zeros((n_frames - T, D), dtype=seq.dtype)
    return np.vstack([seq, pad])

# ------------ tiny on-the-fly augmentation (training only) -----------------
def augment(seq, jitter_std=0.003, frame_drop_prob=0.15):
    """Add small Gaussian noise and randomly drop frames (then re-sample)."""
    if np.random.rand() < frame_drop_prob and seq.shape[0] > FRAMES:
        keep = np.sort(np.random.choice(len(seq), FRAMES, replace=False))
        seq = seq[keep]
    seq = seq + np.random.randn(*seq.shape) * jitter_std
    return seq
# ---------------------------------------------------------------------------

def preprocess_sequences(X, scaler=None, training=False):
    """
    • Sample/pad each sequence (with augmentation if training=True)  
    • Fit/transform StandardScaler on the *flattened* data  
    • Return (X_3d, scaler)
    """
    if training:
        X = [augment(x) for x in X]

    arr = np.stack([sample_sequence(x) for x in X], axis=0)  # (N, F, D)
    flat = arr.reshape(-1, arr.shape[-1])
    if scaler is None:
        scaler = StandardScaler().fit(flat)
    arr = scaler.transform(flat).reshape(arr.shape)
    return arr, scaler

# ─── Tiny-TCN (<20 k params) ───────────────────────────────────────────────
def build_tiny_tcn(n_features,
                   filters=32,
                   kernel=3,
                   dilations=(1, 2),
                   dense_units=(64, 32),
                   dropout=0.1,
                   lr=1e-3):
    K.clear_session()
    inp = Input(shape=(FRAMES, n_features))
    x = inp

    for d in dilations:
        y = SeparableConv1D(filters, kernel, dilation_rate=d,
                            padding='same', use_bias=False)(x)
        y = LayerNormalization()(y)
        y = Activation('relu')(y)
        y = Dropout(dropout)(y)
        # residual path
        if x.shape[-1] != filters:
            x = SeparableConv1D(filters, 1, padding='same', use_bias=False)(x)
        x = Add()([x, y])

    x = GlobalAveragePooling1D()(x)
    for h in dense_units:
        x = Dense(h, activation='relu')(x)
        x = Dropout(dropout)(x)
    out = Dense(1, activation='linear')(x)

    model = tf.keras.Model(inp, out)
    model.compile(optimizer=Adam(lr), loss='mse', metrics=['mae'])
    # safety net – crash if we exceed the budget
    assert model.count_params() <= 20_000, f"Model too big: {model.count_params()} params"
    return model

# ─── Metrics & Plot helpers (unchanged) ────────────────────────────────────
def compute_metrics(y_true, y_pred):
    return (mean_squared_error(y_true, y_pred),
            mean_absolute_error(y_true, y_pred),
            pearsonr(y_true, y_pred)[0])

def plot_pred_vs_true(ax, y_t, y_p, title=None):
    ax.scatter(y_t, y_p, alpha=0.7)
    ax.plot([0, 5], [0, 5], 'k--')
    ax.set_xlim(0,5); ax.set_ylim(0,5)
    ax.set_xlabel('True'); ax.set_ylabel('Pred')
    if title: ax.set_title(title)

def bland_altman(ax, y_t, y_p, title=None):
    diff = y_p - y_t
    mn   = (y_p + y_t) / 2
    md, sd = diff.mean(), diff.std(ddof=1)
    ax.scatter(mn, diff, alpha=0.6)
    ax.axhline(0,  color='k')
    ax.axhline(md, color='b', label=f'mean {md:.2f}')
    ax.axhline(md + 1.96*sd, color='b', ls='--', label='95% CI')
    ax.axhline(md - 1.96*sd, color='b', ls='--')
    ax.set_xlim(0,5); ax.set_ylim(-1.5,1.5)
    if title: ax.set_title(title); ax.legend()

# ─── Cross-validation ──────────────────────────────────────────────────────
def evaluate_cv(X, y, params, n_splits=CV_SPLITS):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_STATE)
    maes = []
    for tr_idx, val_idx in kf.split(X):
        X_tr_raw = [X[i] for i in tr_idx]
        X_val_raw = [X[i] for i in val_idx]

        X_tr, scaler = preprocess_sequences(X_tr_raw, training=True)
        X_val, _     = preprocess_sequences(X_val_raw, scaler)

        model = build_tiny_tcn(X_tr.shape[-1],
                               filters=params['filters'],
                               dense_units=params['dense_units'],
                               dropout=params['dropout'],
                               lr=params['lr'])
        es   = EarlyStopping(patience=ES_PATIENCE, restore_best_weights=True, verbose=0)
        rlrop = ReduceLROnPlateau(factor=0.5, patience=8, min_lr=1e-5, verbose=0)

        model.fit(X_tr, y[tr_idx],
                  validation_data=(X_val, y[val_idx]),
                  epochs=EPOCHS, batch_size=params['batch_size'],
                  callbacks=[es, rlrop], verbose=0)
        _, mae = model.evaluate(X_val, y[val_idx], verbose=0)
        maes.append(mae)
    return np.array(maes)

def grid_search(X, y, grid):
    combos = list(ParameterGrid(grid))
    best_mae, best_params = math.inf, None
    for i, p in enumerate(combos, 1):
        print(f"[{i}/{len(combos)}] testing {p} …")
        maes = evaluate_cv(X, y, p)
        m, s = maes.mean(), maes.std()
        print(f"    MAE {m:.4f} ± {s:.4f}")
        if m < best_mae:
            best_mae, best_params = m, p
    print(f"\nBest params: {best_params}  (MAE {best_mae:.4f})\n")
    return best_params

# ─── Final train / evaluate / save ─────────────────────────────────────────
def train_full(X_raw, y, params):
    X_tr, scaler = preprocess_sequences(X_raw, training=True)
    model = build_tiny_tcn(X_tr.shape[-1],
                           filters=params['filters'],
                           dense_units=params['dense_units'],
                           dropout=params['dropout'],
                           lr=params['lr'])
    es   = EarlyStopping(patience=ES_PATIENCE, restore_best_weights=True, verbose=1)
    rlrop = ReduceLROnPlateau(factor=0.5, patience=8, min_lr=1e-5, verbose=1)

    history = model.fit(X_tr, y,
                        validation_split=0.10,
                        epochs=EPOCHS,
                        batch_size=params['batch_size'],
                        callbacks=[es, rlrop],
                        verbose=1)
    return model, scaler, history

def evaluate(model, scaler, X_raw, y, set_name='Set'):
    X, _ = preprocess_sequences(X_raw, scaler)
    y_p  = model.predict(X, verbose=0)[:,0]
    mse, mae, r = compute_metrics(y, y_p)
    print(f"{set_name:>6} – MSE {mse:.4f} | MAE {mae:.4f} | r = {r:.3f}")
    return y_p

def save(model, scaler, out_dir='models'):
    Path(out_dir).mkdir(exist_ok=True)
    model_path  = str(Path(out_dir, 'squat_scorer.keras'))
    scaler_path = str(Path(out_dir, 'squat_scaler.pkl'))
    model.save(model_path)
    with open(scaler_path, 'wb') as f:
        pickle.dump(scaler, f)
    print(f"\nArtifacts saved:\n  • {model_path}\n  • {scaler_path}")

# ─── Main ------------------------------------------------------------------
if __name__ == '__main__':
    X_all, y_all = load_data(SCORE_PATH, SEQUENCE_DIR)

    # --- tiny 10 % test split
    X_train_raw, X_test_raw, y_train, y_test = train_test_split(
        X_all, y_all, test_size=TEST_SIZE, random_state=RANDOM_STATE)

    best = grid_search(X_train_raw, y_train, param_grid)

    model, scaler, _ = train_full(X_train_raw, y_train, best)
    model.summary()

    print("\n>> Final evaluation")
    y_pred_tr = evaluate(model, scaler, X_train_raw, y_train, 'Train')
    y_pred_te = evaluate(model, scaler, X_test_raw,  y_test,  'Test')

    # quick diagnostics
    fig, ax = plt.subplots(1, 2, figsize=(12,5))
    plot_pred_vs_true(ax[0], y_train, y_pred_tr, 'Train')
    plot_pred_vs_true(ax[1], y_test,  y_pred_te, 'Test')
    plt.tight_layout(); plt.show()

    fig, ax = plt.subplots(1, 2, figsize=(12,5))
    bland_altman(ax[0], y_train, y_pred_tr, 'Train BA')
    bland_altman(ax[1], y_test,  y_pred_te, 'Test  BA')
    plt.tight_layout(); plt.show()

    save(model, scaler)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.model_selection import train_test_split

# 1. Data Preparation (replace this with actual data loading as needed)
np.random.seed(42)
# Simulate dataset: 107 samples, 48 frames, 36 features (3D joint coords per frame), and a squat quality score
X = np.random.randn(107, 48, 36).astype("float32")        # random joint positions
y = np.random.uniform(0, 4, size=(107,)).astype("float32")  # random scores in [0,4]

# Train-test split (e.g., 25% test as a "small test set")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Feature normalization (compute mean/std on training set and apply to both train and test)
feat_mean = X_train.mean(axis=(0,1), keepdims=True)
feat_std  = X_train.std(axis=(0,1), keepdims=True) + 1e-8
X_train = (X_train - feat_mean) / feat_std
X_test  = (X_test  - feat_mean) / feat_std

# 2. Define an Attention layer for sequence pooling
class Attention(layers.Layer):
    def __init__(self):
        super().__init__()
        # Dense layer to compute attention scores for each timestep (tanh activation for non-linearity)
        self.score_layer = layers.Dense(1, activation="tanh")
    def call(self, features):
        # features shape: (batch, timesteps, units)
        # Compute attention scores for each time step
        scores = self.score_layer(features)             # (batch, timesteps, 1)
        weights = tf.nn.softmax(scores, axis=1)         # (batch, timesteps, 1), normalized attention weights
        # Weighted sum of features across timesteps
        context = tf.reduce_sum(features * weights, axis=1)  # (batch, units)
        return context

# 3. Build the model architecture
inputs = layers.Input(shape=(48, 36), name="input_sequence")
# Frame-wise embedding to reduce dimensionality
x = layers.TimeDistributed(layers.Dense(16, activation="relu",
                                       kernel_regularizer=regularizers.l2(1e-4)))(inputs)
x = layers.Dropout(0.2)(x)
# First GRU layer (return sequence for next layer)
x = layers.GRU(50, return_sequences=True, dropout=0.2, recurrent_dropout=0.2,
               kernel_regularizer=regularizers.l2(1e-4),
               recurrent_regularizer=regularizers.l2(1e-4))(x)
# Second GRU layer (return sequence for attention)
x = layers.GRU(34, return_sequences=True, dropout=0.2, recurrent_dropout=0.2,
               kernel_regularizer=regularizers.l2(1e-4),
               recurrent_regularizer=regularizers.l2(1e-4))(x)
# Attention pooling over time steps
x = Attention()(x)
# Optional intermediate dense layer for prediction
x = layers.Dense(10, activation="relu", kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.2)(x)
# Output layer (single score regression)
output = layers.Dense(1, activation="linear", kernel_regularizer=regularizers.l2(1e-4))(x)

model = models.Model(inputs=inputs, outputs=output, name="SquatScoreModel")
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss="mse", metrics=["mae"])

# 4. Model summary and parameter count
model.summary()

# 5. Train the model with early stopping
callbacks = [tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)]
history = model.fit(X_train, y_train,
                    epochs=100, batch_size=16,
                    validation_split=0.2,   # use 20% of training data as validation
                    callbacks=callbacks,
                    verbose=0)  # verbose=0 for brevity; use 1 to see training progress

# 6. Evaluate on the test set
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Test MAE: {test_mae:.4f}")
```
